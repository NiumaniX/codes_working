```python
# 加载相关的包
from WindPy import *
import time
import pandas as pd
w.start()
import collections
import datetime
import openpyxl
from datetime import timedelta
from datetime import datetime
```

```python
#全Apb_lf数据
stock=w.wset("sectorconstituent","date=2021-08-09;sectorid=a001010100000000").Data[1]
df_pb_lf = pd.DataFrame(columns = stock)
#全Ape数据
stock=w.wset("sectorconstituent","date=2021-08-09;sectorid=a001010100000000").Data[1]
df_pe = pd.DataFrame(columns = stock)
```

```python
import datetime
tradedate="20000101"
tradedate_copy = tradedate
end="20000131"
fmt = '%Y%m%d'
tradedate=datetime.datetime.strptime(tradedate,fmt)
end=datetime.datetime.strptime(end,fmt)
delta=datetime.timedelta(days=1)
interval=int((end-tradedate).days)
```

```python
from datetime import datetime
from datetime import timedelta

def add_date(date_str,add_count=1):
    date_list = time.strptime(date_str, "%Y-%m-%d")
    y, m, d = date_list[:3]
    delta = timedelta(days=add_count)
    date_result = datetime(y, m, d) + delta
    date_result = date_result.strftime("%Y-%m-%d")
    print(date_result)
    return date_result
```

```python
date_list = time.strptime(tradedate_copy, "%Y%m%d")
y, m, d = date_list[:3]
add_count=0
while add_count<=interval:
    delta = timedelta(days=add_count)
    date_result = datetime(y, m, d) + delta
    date_result = date_result.strftime("%Y%m%d")
    q = time.strptime(date_result, "%Y%m%d")
    timeStamp = int(time.mktime(q))
    timeArray = time.localtime(timeStamp)
    dateIndex = time.strftime("%Y-%m-%d", timeArray)
    a=date_result
    PB=w.wss(stock,"pb_lf",tradeDate=a).Data[0]
    PE=w.wss(stock,"pe",tradeDate=a,ruleType="10").Data[0]
    df_pb_lf.loc[dateIndex] = PB
    df_pe.loc[dateIndex] = PE
    add_count+=1
```

```python
#指标提取
net_profit=w.wsd(stock, "net_profit_is", "2000-01-01", "2021-08-02", "Period=Q;rptType=1")
oper_rev=w.wsd(stock, "oper_rev", "2000-01-01", "2021-08-02", "Period=Q;rptType=1")
assetsturn=w.wsd(stock, "assetsturn", "2000-01-01", "2021-08-01", "Period=Q")
dupont_assetstoequity=w.wsd(stock, "dupont_assetstoequity", "2000-01-01", "2021-08-02", "Period=Q")
roic=w.wsd(stock, "roic", "2000-01-01", "2021-08-02", "Period=Q")


NET_PROFIT=pd.DataFrame(net_profit.Data,index=stock,columns=net_profit.Times).T
OPER_REV=pd.DataFrame(oper_rev.Data,index=stock,columns=oper_rev.Times).T
营业收入净利率=NET_PROFIT / OPER_REV
ASSETSTURN=pd.DataFrame(assetsturn.Data,index=stock,columns=assetsturn.Times).T
ROA=营业收入净利率*ASSETSTURN
DUPONT_ASSETSTOEQUITY=pd.DataFrame(dupont_assetstoequity.Data,index=stock,columns=dupont_assetstoequity.Times).T
ROE=ROA*DUPONT_ASSETSTOEQUITY
ROIC=pd.DataFrame(roic.Data,index=stock,columns=roic.Times).T

#管理费用
ger_admin_exp=w.wsd(stock, "gerl_admin_exp", "2000-01-01", "2021-08-02", "Period=Q;rptType=1")
GER_ADMIN=pd.DataFrame(ger_admin_exp.Data,index=stock,columns=ger_admin_exp.Times).T
#研发费用
rd_exp=w.wsd(stock, "rd_exp", "2000-01-01","2021-08-02","Period=Q;rptType=1")
RD_EXP=pd.DataFrame(rd_exp.Data,index=stock,columns=rd_exp.Times).T
#销售费用
selling_dist_exp=w.wsd(stock, "selling_dist_exp", "2000-01-01","2021-08-02","Period=Q;rptType=1")
SELLING_DIST_EXP=pd.DataFrame(selling_dist_exp.Data,index=stock,columns=selling_dist_exp.Times).T
#财务费用
fin_exp=w.wsd(stock, "fin_exp_is", "2000-01-01","2021-08-02","Period=Q;rptType=1")
FIN_EXP=pd.DataFrame(fin_exp.Data,index=stock,columns=fin_exp.Times).T
#销售毛利率
grossprofitmargin=w.wsd(stock, "grossprofitmargin", "2000-01-01", "2021-08-02", "Period=Q")
GROSSPROFITMARGIN=pd.DataFrame(grossprofitmargin.Data,index=stock,columns=grossprofitmargin.Times).T
#资产负债率
debttoassets=w.wsd(stock, "debttoassets", "2000-01-01", "2021-08-02", "Period=Q")
DEBTTOASSETS=pd.DataFrame(debttoassets.Data,index=stock,columns=debttoassets.Times).T
#存货周转率
invturn=w.wsd(stock, "invturn", "2000-01-01", "2021-08-02", "Period=Q")
INVTURN=pd.DataFrame(invturn.Data,index=stock,columns=invturn.Times).T
```
